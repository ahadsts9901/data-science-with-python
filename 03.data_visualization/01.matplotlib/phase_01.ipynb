# Import libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Basic Plot

```python
x = np.linspace(0,5,11)
x
y = x ** 2
y
```

```python
plt.plot(x,y)
plt.title("title goes here")
plt.xlabel("x axis custom title")
plt.ylabel("y axis custom title")
```

# Creating Subplots

```python
plt.subplot(2, 2, 1)
plt.plot(x,y)
plt.subplot(2, 2, 3)
plt.plot(y,x)
plt.subplot(2, 2, 2)
plt.plot(x,x)
plt.subplot(2, 2, 4)
plt.plot(y,y)
```

# Object Oriented Way to move sub_plots

```python
fig = plt.figure(figsize=(8,5), dpi=100)
axis1 = fig.add_axes([0.1, 0.1, 1,1])
axis1.plot(x,y)
axis2 = fig.add_axes([0.8, 0.5, 0.2,0.2])
axis2.plot(x,y)
```

# Types Of Plot

```python
plt.scatter(x,y)
```

```python
plt.hist(y)
```

```python
plt.boxplot(x)
```

# Saving Plots

```python
fig = plt.figure()
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])
ax.plot(x, y)
plt.savefig("basic_plot")
```
